```python
import tripodpy
import matplotlib.pyplot as plt
import numpy as np
from dustpy import constants as c
```

```python
# 1. Crear el objeto de simulación (el cascarón)
sim = tripodpy.Simulation()

# 2. INICIALIZAR LA SIMULACIÓN (Paso crucial para cargar sim.t, sim.grid, etc.)
sim.initialize()

# 3. Definir los snapshots con espaciado logarítmico
# Guardamos 6 snapshots entre 1,000 (10^3) y 10,000 (10^4) años
sim.t.snapshots = np.logspace(3, 4, num=6, base=10) * c.year

# 4. Correr la simulación
print("Iniciando la simulación...")
sim.run()
print("¡Simulación terminada!")

# 5. Visualizar los resultados
fig, ax = plt.subplots(figsize=(8, 5))

ax.loglog(sim.grid.r / c.au, sim.gas.Sigma, label='Gas', color='blue', lw=2)
ax.loglog(sim.grid.r / c.au, sim.dust.Sigma, label='Polvo', color='red', lw=2)

ax.set_xlabel('Radio [au]')
ax.set_ylabel('Densidad Superficial [g/cm²]')
ax.set_title('Perfil de Densidad del Disco Liso (Modelo Estándar)')
ax.legend()
ax.grid(True, which="both", ls="--", alpha=0.5)

plt.show()
```
